```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# LOAD DATA
# -------------------------------
trades = pd.read_csv("historical_data.csv")
sentiment = pd.read_csv("fear_greed.csv")

# -------------------------------
# CLEAN COLUMN NAMES (IMPORTANT)
# -------------------------------
trades.columns = trades.columns.str.strip().str.lower().str.replace(" ", "_")
sentiment.columns = sentiment.columns.str.strip().str.lower().str.replace(" ", "_")

print("Trades Columns:", trades.columns)
print("Sentiment Columns:", sentiment.columns)

# -------------------------------
# FIND TIME COLUMN AUTOMATICALLY
# -------------------------------
time_col = None
for col in trades.columns:
    if 'time' in col:
        time_col = col
        break

# Fallback (important for your dataset)
if time_col is None:
    for col in trades.columns:
        if 'stamp' in col:
            time_col = col
            break

if time_col is None:
    raise Exception("❌ No time column found in trades dataset")

print("Using time column:", time_col)

# -------------------------------
# CREATE DATE COLUMN
# -------------------------------
trades['date'] = pd.to_datetime(trades[time_col], errors='coerce').dt.strftime('%Y-%m-%d')

# Fix sentiment date column
if 'date' not in sentiment.columns:
    raise Exception("❌ Sentiment dataset missing date column")

sentiment['date'] = pd.to_datetime(sentiment['date'], errors='coerce').dt.strftime('%Y-%m-%d')

# -------------------------------
# MERGE DATA
# -------------------------------
df = pd.merge(trades, sentiment, on='date', how='inner')

print("Merged Shape:", df.shape)

if df.empty:
    raise Exception("❌ Merge failed → No matching dates")

# -------------------------------
# FIND IMPORTANT COLUMNS
# -------------------------------
pnl_col = None
for col in df.columns:
    if 'pnl' in col:
        pnl_col = col
        break

class_col = None
for col in df.columns:
    if 'class' in col:
        class_col = col
        break

if pnl_col is None or class_col is None:
    raise Exception("❌ Required columns not found")

print("Using PnL column:", pnl_col)
print("Using Classification column:", class_col)

# -------------------------------
# CLEAN DATA
# -------------------------------
df = df.dropna(subset=[pnl_col, class_col])
df[pnl_col] = pd.to_numeric(df[pnl_col], errors='coerce')

# -------------------------------
# ANALYSIS
# -------------------------------
print("\nAverage Profit:")
print(df.groupby(class_col)[pnl_col].mean())

print("\nTrade Count:")
print(df[class_col].value_counts())

# -------------------------------
# VISUALIZATION
# -------------------------------
sns.set()

# Profit distribution
plt.figure()
sns.boxplot(x=class_col, y=pnl_col, data=df)
plt.title("Profit Distribution by Sentiment")
plt.show()

# Average profit
plt.figure()
df.groupby(class_col)[pnl_col].mean().plot(kind='bar')
plt.title("Average Profit by Sentiment")
plt.show()
```
